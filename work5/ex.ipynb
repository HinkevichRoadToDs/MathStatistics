2 задача: Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные
автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из
n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

```python
mu = 17
alpha = 0.05
n = 100
xmean = 17.5
var = 4
# H0 : xmean = mu
# H1 : xmean > mu (правосторонний критерий)
Z = (xmean - mu) / (2 / np.sqrt(n))
# найду квантиль порядка 0.95
print(f'z критическое = {stats.norm.ppf(0.95)}, z наблюдаемое = {Z}\n'
      'Отвергаю нулевую гипотезу, так как z наблюдаемое сильно больше отклонилось от '
      'среднего ГС, чем z критический критическое')
```

Задача 3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья
составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна
99%? (Провести двусторонний тест.)

```python
mu = 200
x = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])  # x ~ N(xmean,sd)
n = x.shape[0]
xmean = x.sum() / n
alpha = 0.01
sd = np.square(x - xmean).sum() / n - 1
# H0 : xmean = mu
# H1 : xmean != mu
t_emp = (xmean - mu) / (sd / np.sqrt(n))
# по таблице t-распределения квантили порядка alpha/2 и 1 - alpha/2 равны -3,25 и 3,25
t1 = stats.t.ppf(alpha / 2, df=9)
t2 = stats.t.ppf(1 - alpha / 2, df=9)
print('t1:', t1, 't-статистика:', t_emp, 't2:', t2, 'Принимаю нулевую гипотезу, так как t-статистика не выходит за '
                                                    'критические границы')
```
